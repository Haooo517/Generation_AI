```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

我決定製作一個很有名的函數圖形：**愛心函數圖形**(長這樣→♡)  <br>
不過使用的不是傳說中的**笛卡兒公式**<br>
而是$\left(x^{2}+y^{2}-1\right)^{3}=x^{2}y^{3}$ <br>
呈現的效果能比前者更好 <br>
因此先定義一個愛心函數的函式為 `heart()`

```python
def heart(t):
  x = 16 * np.sin(t)**3
  y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)
  return x, y
```

接著使用 `np.linspace` 生成一系列的 `t` 值 <br>
並透過 `heart()` 函數計算對應的 `x` 和 `y` 值

```python
t = np.linspace(0, 2 * np.pi, 500)
x, y = heart(t)
```

再來用 `plt` 實現以下步驟：<br>
1. **繪製圖形**: 使用 `plt.plot` 繪製心形線，並設定顏色、線條寬度等屬性<br>
2. **設定圖形標題和軸標籤**: 使用 `plt.title`、`plt.xlabel` 和 `plt.ylabel` 設定圖形標題和軸標籤，並調整字體大小
3. **顯示網格線**: 使用 `plt.grid` 顯示網格線，並調整樣式和透明度<br>
4. **設定軸比例**: 使用 `plt.axis('equal')` 設定 x 和 y 軸比例相等，使心形線顯示更完整<br>
5. **填充圖形**: 使用 `plt.fill_between` 填充心形線內部，並調整顏色和透明度<br>
6. **新增文字註解**: 使用 `plt.text` 新增文字註解，並設定位置、字體大小、顏色等屬性

```python
plt.plot(x, y, color='red', linewidth=2)
plt.title('The Heart Line', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axis('equal')
plt.fill_between(x, y, color='pink', alpha=0.5)
plt.show()
```
